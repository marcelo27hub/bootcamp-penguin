```python
cantidad_de_claves= int(input("ingrese la cantidad de valores: "))
tarjeta_personal= {}
for i in range(cantidad_de_claves):
    clave= input("ingrese la clave: ")
    tarjeta_personal[clave]= input("ingrese el valor de la clave: ")
print(tarjeta_personal)

while True:
    usuario= input("ingrese que quiere hacer con la tarjeta personal (crear/modificar/eliminar/salir del programa): ").lower()
    if usuario== "crear":
        clave_nuevo= input("ingrese una clave a agg: ")
        tarjeta_personal[clave_nuevo]= input("ingrese el valor de la clave: ")
        print(tarjeta_personal)    
    elif usuario== "modificar":
        clave_a_modificar= input("ingrese la clave a modificar: ")
        if clave_a_modificar in tarjeta_personal:
            tarjeta_personal[clave_a_modificar]= input("ingrese el nuevo valor de la clave modificado: ")
            print(tarjeta_personal)
        else:
            print("No existe esa clave")   
    elif usuario== "eliminar":
        clave_a_eliminar = input("ingrese la clave a eliminar: ")
        if clave_a_eliminar in tarjeta_personal:
            del tarjeta_personal[clave_a_eliminar]
            print(tarjeta_personal)
        else:
            print("No existe esa clave")
    elif usuario== "salir del programa":
        print("saliendo del programa!")
        break
    else:
        print("la opcion no existe")    
print(tarjeta_personal)        
                    
            
        
             
        
           
```

```python
agenda= {"usuario1":{"nombre": "marcelo", 
                     "apellido": "valdez"},
         "usuario2":{"nombre": "valentina",
                     "apellido": "alavarenga"}}
print("""      1.mostrar los contactos
      2. crear los contactos
      3. modificar los contactos
      4. salir del programa""")
while True:
    usuario_opcion= int(input("ingrese una de las opciones(1,2 o 3): "))
    if usuario_opcion == 1:
        print(agenda.keys())
        contacto_usuario= input("ingrese el contacto que quiere ver: ").lower()
        if contacto_usuario in agenda:
            print(agenda[contacto_usuario])
        else:
            print("no existe ese usuario en la agenda")
    elif usuario_opcion == 2:
        nuevo_usuario = input("Ingrese un nuevo usuario ej:(usuario3,4 u otro): ").lower()
        nombre = input("Ingrese el nombre: ").lower()
        apellido = input("Ingrese el apellido: ").lower()
        agenda[nuevo_usuario] = {
        "nombre": nombre,
        "apellido": apellido
        }
        print(agenda)
    elif usuario_opcion == 3:
        print(agenda.keys())
        contacto_a_modificar= input("ingrese un contacto a modificar: ").lower()
        if contacto_a_modificar in agenda:
            campo= input("ingrese que quiere modificar (nombre/apellido): ").lower()
            if campo in agenda[contacto_a_modificar]:
                valor_nuevo_del_contacto= input("ingrese el valor del campo: ").lower()
                agenda[contacto_a_modificar][campo]= valor_nuevo_del_contacto
                print(agenda[contacto_a_modificar])
            else:
                print("no existe ese campo")
                
    elif usuario_opcion== 4:
        print("saliendo del programa")
        break
    else:
        print("ingrese una de las 4 opciones!")

print(agenda)                            
            
        
                
    
            

    
```
